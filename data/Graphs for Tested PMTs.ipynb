```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
#Load Data / set up x axis
x = np.array([1,2,3,4,5,6,7,8])
ave_trig_rate = np.loadtxt('averagescintillatorcounts.csv',delimiter = ',')
colors = cm.rainbow(np.linspace(0, 1, len(x)))
```

```python
#Plot the Data
fig = plt.figure()
plt.title('Scintillator PMT Average Trigger Rates',)
plt.xlabel('PMT')
plt.ylabel('Average PMT Trigger (Hz)')
plt.plot( x[0], ave_trig_rate[0], 'ro', label = 'PMT T-BOT N')
plt.plot( x[1], ave_trig_rate[1], 'bo', label = 'PMT T-TOP N')
plt.plot( x[2], ave_trig_rate[2], 'ko', label = 'PMT T-BOT S')
plt.plot( x[3], ave_trig_rate[3], 'mo', label = 'PMT T-TOP S')
plt.plot( x[4], ave_trig_rate[4], 'co', label = 'PMT H-TOP N')
plt.plot( x[5], ave_trig_rate[5], 'yo', label = 'PMT H-TOP S')
plt.plot( x[6], ave_trig_rate[6], 'go', label = 'PMT H-BOT N')
plt.plot( x[7], ave_trig_rate[7], 'r^', label = 'PMT H-BOT S')
plt.legend(loc='upper left')
```
